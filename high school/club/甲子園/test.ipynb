```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSVファイルを読み込む
df = pd.read_csv('5meiden.csv')

# 球種ごとの成績を集計する
kind_performance = df.groupby('kind')['result'].value_counts().unstack().fillna(0)

# コースごとの成績を集計する
course_performance = df.groupby('course')['result'].value_counts().unstack().fillna(0)

# 成績をグラフにする
plt.figure(figsize=(10, 5))
sns.heatmap(kind_performance, annot=True, fmt='g')
plt.title('kind')
plt.show()

plt.figure(figsize=(10, 5))
sns.heatmap(course_performance, annot=True, fmt='g')
plt.title('course')
plt.show()
```
